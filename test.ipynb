Simple YOLO starter code

```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo26n.pt")  # load an official model

# Predict with the model
results = model("https://ultralytics.com/images/bus.jpg")  # predict on an image

# Access the results
for result in results:
    xywh = result.boxes.xywh  # center-x, center-y, width, height
    xywhn = result.boxes.xywhn  # normalized
    xyxy = result.boxes.xyxy  # top-left-x, top-left-y, bottom-right-x, bottom-right-y
    xyxyn = result.boxes.xyxyn  # normalized
    names = [result.names[cls.item()] for cls in result.boxes.cls.int()]  # class name of each box
    confs = result.boxes.conf  # confidence score of each box
```

User Display Mode

```python
import cv2
from ultralytics import YOLO

model = YOLO("yolo26n.pt")  

camera = cv2.VideoCapture(0)

if not camera.isOpened():
    print("Error: Could not open video stream.")
    exit()

print("Live video stream active. Press 'q' inside the window to exit.")

# Runs a continuous loop
while True:
    ret, frame = camera.read()
    if not ret:
        print("Error: Failed to grab frame.")
        break

    results = model(frame, stream=True)

    for result in results:
        boxes = result.boxes
        for box in boxes:
            # Get integer coordinates for drawing
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            # Extract confidence and class ID
            conf = float(box.conf[0])
            class_id = int(box.cls[0])
            label = f"{result.names[class_id]} {conf:.2f}"

            # Draw the bounding box and label directly onto the live frame
            # cv2.rectangle syntax: (image, top-left, bottom-right, color_bgr, thickness)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    cv2.imshow("YOLO Live Detection Feed", frame)

    # Check if the user pressed the 'q' key to quit
    # waitKey(1) introduces a 1ms delay required for OpenCV window rendering
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Clean up and release the hardware resources
camera.release()
cv2.destroyAllWindows()
```
